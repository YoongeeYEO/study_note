# 028. Median of Two Sorted Arrays


[문제 028](https://neetcode.io/problems/median-of-two-sorted-arrays)

- You are given two integer arrays nums1 and nums2 of size m and n respectively, where each is sorted in ascending order. Return the median value among all elements of the two arrays.

- Your solution must run in O(log(m+n)) time.

Constraints:
- nums1.length == m
- nums2.length == n
- 0 <= m <= 1000
- 0 <= n <= 1000
- -10^6 <= nums1[i], nums2[i] <= 10^6


## Solution
- nums1과 nums2를 오름차순으로 정렬하여 merge
- 합친 리스트의 길이가 짝수인 경우 mid1 + mid2의 평균이 중앙값
- 합친 리스트의 길이가 홀수인 경우 리스트의 중앙값의 값을 반환


```python
from typing import List

class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        merged_list = sorted(nums1 + nums2)
        total_len = len(merged_list)

        if total_len % 2 == 0:
            mid1 = total_len // 2 - 1
            mid2 = mid1 + 1
            median = (merged_list[mid1] + merged_list[mid2]) / 2
        else:
            mid = total_len // 2
            median = merged_list[mid]

        return median
```

```python
nums1 = [1,2]
nums2 = [3]

solution = Solution()
solution.findMedianSortedArrays(nums1, nums2)
```
