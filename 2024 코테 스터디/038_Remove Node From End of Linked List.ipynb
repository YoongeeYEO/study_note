# 038. Remove Node From End of Linked List


[문제 038](https://neetcode.io/problems/remove-node-from-end-of-linked-list)

- You are given the beginning of a linked list head, and an integer n.

- Remove the nth node from the end of the list and return the beginning of the list.

Constraints:
- The number of nodes in the list is sz.
- 1 <= sz <= 30
- 0 <= Node.val <= 100
- 1 <= n <= sz

## Solution
- 원래 리스트 앞에 빈 더미 리스트 생성
- 더미 노드에서 시작하는 left와 실제 리스트 앞에서 시작하는 right를 생성해준다.
- n만큼 left와 right를 이동시킨다. right의 값이 none이 될때까지..이때 left는 삭제할 직전 노드를 가리키게된다.
- 원래 리스트에서 left의 다음 노드를 삭제한 후 원래 리스트를 반환

```python
from typing import Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        dummy = ListNode(0, head)
        left = dummy
        right = head

        while n > 0 and right:
            right = right.next
            n -= 1

        while right:
            left = left.next
            right = right.next

        left.next = left.next.next
        return dummy.next
```

```python
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4))))
n = 2
solution = Solution()
solution.removeNthFromEnd(head, n)
```
