# 012. Three Integer Sum

[문제 012](https://neetcode.io/problems/three-integer-sum)

- Given an integer array nums, return all the triplets [nums[i], nums[j], nums[k]] where nums[i] + nums[j] + nums[k] == 0, and the indices i, j and k are all distinct.

- The output should not contain any duplicate triplets. You may return the output and the triplets in any order.

Constraints:
- 3 <= nums.length <= 1000
- -10^5 <= nums[i] <= 10^5


## Solutions
- for문 3개 돌려서 nums[i]+nums[j]+nums[k] == 0 이 되는 값을 찾음.
- 찾은 후 중복을 제거하기 위해 set()에 넣어준다. set에 넣어주려면 tuple 형태로 바꿔서 넣어줘야하는듯...
- set에 넣어진 값을 리스트로 변환하여 출력

```python
from typing import List

class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        answer = set()
        nums.sort()
        for i in range(len(nums)):
            for j in range(i+1,len(nums)):
                for k in range(j+1,len(nums)):
                    if nums[i]+nums[j]+nums[k] == 0:
                        answer.add(tuple([nums[i],nums[j],nums[k]]))
        return [list(i) for i in answer]
```

```python
nums = [0,0,0]

s = Solution()
s.threeSum(nums)
```

```python
nums = [0,1,1]

s = Solution()
s.threeSum(nums)
```

```python
nums=[-1,0,1,2,-1,-4]

s = Solution()
s.threeSum(nums)
```
