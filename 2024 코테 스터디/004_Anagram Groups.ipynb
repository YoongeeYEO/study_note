# 004. Anagram Groups

[문제 004](https://neetcode.io/problems/anagram-groups)

- Given an array of strings strs, group all anagrams together into sublists. You may return the output in any order.

- An anagram is a string that contains the exact same characters as another string, but the order of the characters can be different.

Constraints:

- 1 <= strs.length <= 1000.
- 0 <= strs[i].length <= 100
- strs[i] is made up of lowercase English letters.

## Solutions

- Dictionary의 key값을 가지는 빈 리스트를 생성
- 문자열을 정렬했을때 같은 순서로 정렬되어 동일한 문자열이 될때 Anagram이 되므로 문자열 정렬.
- 정렬된 문자열을 key값으로 가지는 dictionary에 저장..같은 anagram의 경우 같은 key값을 가지는 리스트 내에 함께 저장된다.
- dictionary의 모든 값(리스트)을 반환

```python
from collections import defaultdict
from typing import List

class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        anagrams = defaultdict(list)

        for s in strs:
            sorted_str = ''.join(sorted(s))
            anagrams[sorted_str].append(s)

        return list(anagrams.values())
```

```python
# Example
strs = ["act","pots","tops","cat","stop","hat"]
solution = Solution()
result = solution.groupAnagrams(strs)
print(result)
```
