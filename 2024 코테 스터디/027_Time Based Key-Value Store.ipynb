# 027. Time Based Key-Value Store


[문제 027](https://neetcode.io/problems/time-based-key-value-store)

- Implement a time-based key-value data structure that supports:

  - Storing multiple values for the same key at specified time stamps
  - Retrieving the key's value at a specified timestamp
- Implement the TimeMap class:

  - TimeMap() Initializes the object.
  - void set(String key, String value, int timestamp) Stores the key key with the value value at the given time timestamp.
  - String get(String key, int timestamp) Returns the most recent value of key if set was previously called on it and the most recent timestamp for that key prev_timestamp is less than or equal to the given timestamp (prev_timestamp <= timestamp). If there are no values, it returns "".
- Note: For all calls to set, the timestamps are in strictly increasing order.

Constraints:
- 1 <= key.length, value.length <= 100
- key and value only include lowercase English letters and digits.
- 1 <= timestamp <= 1000


## Solution
- def __init__(self)
  - 딕셔너리 생성
- def set(self, key: str, value: str, timestamp: int)
  - 딕셔너리에 입력된 key 값에 대응하는 value 추가
- def get(self, key: str, timestamp: int)
  - key값이 딕셔너리에 없는 경우 "" 반환
  - 해당 키에 대한 (timestamp, value) 리스트를 가져옴
  - 특정 키와 주어진 시간(timestamp)에 대해 가장 최근 값을 반환
  - 이진 탐색(binary search)을 사용해, 주어진 시간보다 작거나 같은 가장 큰 timestamp를 찾음


```python
from typing import List

class TimeMap:

    def __init__(self):
        self.timemap_dict = {}

    def set(self, key: str, value: str, timestamp: int) -> None:
        if key not in self.timemap_dict:
            self.timemap_dict[key] = []
        self.timemap_dict[key].append((timestamp, value))

    def get(self, key: str, timestamp: int) -> str:
        if key not in self.timemap_dict:
            return ""

        values = self.timemap_dict[key]
        left, right = 0, len(values) - 1
        result = ""

        while left <= right:
            mid = (left + right) // 2
            if values[mid][0] <= timestamp:
                result = values[mid][1]
                left = mid + 1
            else:
                right = mid - 1

        return result
```

```python
timeMap = TimeMap();
timeMap.set("alice", "happy", 1);
```

```python
timeMap.get("alice", 1)
```

```python
timeMap.get("alice", 2)
```

```python
timeMap.set("alice", "sad", 3)
```

```python
timeMap.get("alice", 3)
```
