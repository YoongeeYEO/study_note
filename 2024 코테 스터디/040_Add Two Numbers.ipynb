# 040. Add Two Numbers


[문제 040](https://neetcode.io/problems/add-two-numbers)

- You are given two non-empty linked lists, l1 and l2, where each represents a non-negative integer.

- The digits are stored in reverse order, e.g. the number 123 is represented as 3 -> 2 -> 1 -> in the linked list.

- Each of the nodes contains a single digit. You may assume the two numbers do not contain any leading zero, except the number 0 itself.

- Return the sum of the two numbers as a linked list.

Constraints:
- 1 <= l1.length, l2.length <= 100.
- 0 <= Node.val <= 9

## Solution
- 빈 더미 리스트 생성
- l1과 l2값, 그리고 올릴 수를 저장하는 up값을 모두 더한 후 10으로 나누고 나머지값을 더미 리스트에 저장
- l1,l2값이 None이 될때까지 l1, l2값을 한칸씩 옮기면서 동일하게 반복

```python
from typing import Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def addTwoNumbers(self, l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
        dummy_head = ListNode(0)
        current = dummy_head
        up = 0

        while l1 or l2 or up:
            val1 = l1.val if l1 else 0
            val2 = l2.val if l2 else 0


            total = val1 + val2 + up
            up = total // 10
            current.next = ListNode(total % 10)


            current = current.next
            if l1: l1 = l1.next
            if l2: l2 = l2.next

        return dummy_head.next
```

```python
l1 = ListNode(1, ListNode(2, ListNode(3)))
l2 = ListNode(4, ListNode(5, ListNode(6)))

solution = Solution()
result = solution.addTwoNumbers(l1, l2)

while result:
    print(result.val, end=" -> ")
    result = result.next
```
