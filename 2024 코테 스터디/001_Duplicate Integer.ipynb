# 001. Duplicate Integer

- Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

- 정수 배열 번호가 주어지면 배열에 두 번 이상 값이 나타나면 true로 반환하고 그렇지 않으면 false로 반환하는 문제.

```python
from typing import List

class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        return len(nums) != len(set(nums))
```

Solution 이라는 클래스 내에 hasDuplicate 함수를 정의하고,

정수형 리스트 형태인 nums를 입력으로 받아 len(nums) 길이와 len(set(nums)) 길이를 비교하여 다를경우 nums 리스트 내 중복이 있다는 의미이므로 True 반환, 그렇지 않으면 False

```python
# Example

nums = [1, 2, 3, 3]

# Create an instance of the Solution class
solution_instance = Solution()

# Call the hasDuplicate method on the instance
solution_instance.hasDuplicate(nums)
```

```python
# Example 2

nums2=[1,6,7,8,9]

solution_instance.hasDuplicate(nums2)
```
