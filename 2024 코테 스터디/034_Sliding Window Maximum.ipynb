# 034. Sliding Window Maximum


[문제 034](https://neetcode.io/problems/permutation-string)

- You are given an array of integers nums and an integer k. There is a sliding window of size k that starts at the left edge of the array. The window slides one position to the right until it reaches the right edge of the array.

- Return a list that contains the maximum element in the window at each step.

Constraints:
- 1 <= nums.length <= 1000
- -1000 <= nums[i] <= 1000
- 1 <= k <= nums.length

## Solution
- 주어진 k값을 활용해서 중복 for문을 활용하여 리스트 nums 내 부분 리스트 값들을 비교하고 가장 큰 값들만 window_answer에 저장
- for문이 끝나면 최종 window_answer를 반환

```python
from typing import List

class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        window_answer = []
        for i in range (len(nums)-k+1):
            max_i = nums[i]
            for j in range(i,i+k):
                max_i = max(max_i, nums[j])
            window_answer.append(max_i)
        return window_answer
```

```python
nums = [1,2,1,0,4,2,6]
k = 3

solution = Solution()
solution.maxSlidingWindow(nums, k)
```
