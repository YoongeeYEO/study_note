# 014. Trapping rain water

[문제 014](https://neetcode.io/problems/trapping-rain-water)

- You are given an array non-negative integers heights which represent an elevation map. Each value heights[i] represents the height of a bar, which has a width of 1.

- Return the maximum area of water that can be trapped between the bars.

public (1).avif

Constraints:
- 1 <= height.length <= 1000
- 0 <= height[i] <= 1000

## Solutions
- 왼쪽과 오른쪽에서 각각 최대 높이를 추적하고, 이를 사용해 각 위치에서 갇힌 물의 양을 결정
- 만약 left_max가 right_max보다 작으면, left 포인터를 오른쪽으로 이동. 그리고 left_max를 업데이트하고, left_max에서 height[left]를 뺀 값을 trapped_water에 더해 현재 위치에서 갇힌 물의 양을 계산
- right_max가 left_max보다 작거나 같다면, right 포인터를 왼쪽으로 이동. right_max를 업데이트하고, right_max에서 height[right]를 뺀 값을 trapped_water에 더해 현재 위치에서 갇힌 물의 양을 계산한다.

```python
from typing import List

class Solution:
    def trap(self, height: List[int]) -> int:
        if not height:
            return 0

        left, right = 0, len(height) - 1
        left_max, right_max = height[left], height[right]
        trapped_water = 0

        while left < right:
            if left_max < right_max:
                left += 1
                left_max = max(left_max, height[left])
                trapped_water += max(0, left_max - height[left])
            else:
                right -= 1
                right_max = max(right_max, height[right])
                trapped_water += max(0, right_max - height[right])

        return trapped_water
```

```python
# Example
height = [0,2,0,3,1,0,1,3,2,1]

s = Solution()
s.trap(height)
```
