2022년 5월 sales의 중앙값을 구하시오

데이터셋 : basic2.csv

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/basic2.csv')

df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'])

df.info()
```

```python
#Datetime -> year / month / day로 쪼개는 법!!

df['year']=df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['date']=df['Date'].dt.day

df.head()
```

```python
cond1 = df['year']==2022
cond2 = df['month']==5

answer = df[cond1&cond2]['Sales'].median()
print(answer)
```
