# 1-1
* age 컬럼의 3사분위수와 1사분위수의 차를 절대값으로 구하고, 소수점 버려서, 정수로 출력하세요

```python
import pandas as pd
df = pd.read_csv('/content/basic1.csv')

df.head()
```

```python
df.info()
```

```python
q1 = df['age'].quantile(0.25)
q3 = df['age'].quantile(0.75)

diff = abs(q1-q3)
print(diff)
```

```python
answer = int(diff.round(0))
print(answer)
```

# 1-2
* (loves반응+wows반응)/(reactions반응) 비율이 0.4보다 크고 0.5보다 작으면서, type 컬럼이 'video'인 데이터의 갯수를 구하세요

```python
df2 = pd.read_csv('/content/fb.csv')
df2.head()
```

```python
df2.info()
```

```python
condition1 = (df2['loves']+df2['wows'])/df2['reactions'] > 0.4
condition2 = (df2['loves']+df2['wows'])/df2['reactions'] < 0.5
condition3 = df2['type'] == 'video'

answer = len(df2[condition1 & condition2 & condition3])

print(answer)
```

# 1-3
* date_added가 2018년 1월 이면서 country가 United Kingdom 단독 제작인 데이터의 갯수를 구하세요

```python
df3 = pd.read_csv('/content/nf.csv')
df3.head()
```

```python
df3['date_added'] = pd.to_datetime(df3['date_added'])

df3['date_added'].head()
```

```python
df3['year'] = df3['date_added'].dt.year
df3['month'] = df3['date_added'].dt.month
df3['day'] = df3['date_added'].dt.day
```

```python
df3.head()
```

```python
cond1 = df3['year'] == 2018
cond2 = df3['month'] == 1
cond3 = df3['country'] == 'United Kingdom'

answer = len(df3[cond1 & cond2 & cond3])
print(answer)
```
