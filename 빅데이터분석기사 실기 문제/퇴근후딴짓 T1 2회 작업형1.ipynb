* 데이터셋(basic1.csv)의 'f5' 컬럼을 기준으로 상위 10개의 데이터를 구하고,
* 'f5'컬럼 10개 중 최소값으로 데이터를 대체한 후,
* 'age'컬럼에서 80 이상인 데이터의'f5 컬럼 평균값 구하기

```python
import pandas as pd
df = pd.read_csv('/content/basic1.csv')
df.head()
```

```python
df.info()
```

```python
df = df.sort_values(by='f5', ascending=False)
df.head()
```

```python
a = df['f5'].min()
print(a)
```

```python
df.iloc[:10,7:]=a
df
```

# 문제2
* 데이터셋(basic1.csv)의 앞에서 순서대로 70% 데이터만 활용해서,
* 'f1'컬럼 결측치를 중앙값으로 채우기 전후의 표준편차를 구하고
* 두 표준편차 차이 계산하기

```python
df1 = pd.read_csv('/content/basic1.csv')
df1.head()
```

```python
df1.info()
```

```python
df1=df1.iloc[:int(100*0.7),:]
df1.info()
```

```python
# f1 컬럼 결측치 채우기 전 표준편차
a = df1['f1'].std()
print(a)
```

```python
# f1 컬럼 결측치 중앙값으로 채우기

med = df1['f1'].median()

df1['f1'] = df1['f1'].fillna(med)

# 잘 채워졌는지 확인
df1.info()
```

```python
b = df1['f1'].std()
print(b)
```

```python
answer = a - b
print(answer)
```

# 문제3
* 데이터셋(basic1.csv)의 'age'컬럼의 이상치를 더하시오
* 단, 평균으로부터 '표준편차*1.5'를 벗어나는 영역을 이상치라고 판단함

```python
df3 = pd.read_csv('/content/basic1.csv')
df3.info()
```

```python
condition1 = df3['age'] < df3['age'].mean() - 1.5*df3['age'].std()
condition2 = df3['age'] > df3['age'].mean() + 1.5*df3['age'].std()
```

```python
new_df3 = df3[condition1 | condition2]

answer = new_df3['age'].sum()
print(answer)
```
