    (최종값은 소수점 반올림 후 정수 출력)
* 데이터셋 : basic2.csv

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/basic2.csv')
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
df.head()
```

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day

df.head()
```

```python
for i in range(0,len(df)):
    if df['Events'][i]==1:
        df['Sales'][i]=df['Sales'][i]*0.8

df.head()
```

```python
a = df[df['year']==2022].groupby('month')['Sales'].sum().max()
print(a)
```

```python
b = df[df['year']==2023].groupby('month')['Sales'].sum().max()
print(b)
```

```python
answer = int(np.round(abs(a-b),0))
print(answer)
```
