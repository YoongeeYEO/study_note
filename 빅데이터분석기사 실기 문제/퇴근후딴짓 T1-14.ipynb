```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/drive/MyDrive/빅분기 실기 연습문제/basic1.csv')
df.head()
```

```python
df_new = pd.DataFrame(df.groupby(['city','f4'])['f5'].mean())

df_new
```

```python
df_new = df_new.sort_values(by='f5', ascending=False).head(7)

answer = np.round(df_new['f5'].sum(),2)

print(answer)
```
