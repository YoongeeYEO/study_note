* 주어진 데이터에서 2022년 5월 주말과 평일의 sales컬럼 평균값 차이를 구하시오 (소수점 둘째자리까지 출력, 반올림)
* 데이터셋 : basic2.csv

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/basic2.csv')
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df.info()
```

```python
#dt.dayofweek : 월화수목금토일 -> 0~6 숫자로 변환

df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['dayofweek'] = df['Date'].dt.dayofweek

df.head()
```

```python
cond1 = df['year']==2022
cond2 = df['month']==5 
cond3 = df['dayofweek']>=5
cond4 = df['dayofweek']<5

a = df[cond1&cond2&cond3]['Sales'].mean()
b = df[cond1&cond2&cond4]['Sales'].mean()

answer = np.round(a-b,2)
print(answer)
```
