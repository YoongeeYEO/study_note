#T1-15. 슬라이싱 & 조건 Expected Question

주어진 데이터 셋에서 age컬럼 상위 20개의 데이터를 구한 다음 

f1의 결측치를 중앙값으로 채운다.

그리고 f4가 ISFJ와 f5가 20 이상인 f1의 평균값을 출력하시오!

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/basic1.csv')
df.head()
```

```python
df_new = df.sort_values(by='age', ascending=False).head(20)

df_new
```

```python
df_new['f1'] = df_new['f1'].fillna(df_new['f1'].median())
```

```python
cond1 = df_new['f4']=='ISFJ'
cond2 = df_new['f5']>=20

answer = df_new[cond1&cond2]['f1'].mean()

print(answer)
```
