# 문제1
* Iris 데이터 세트에서 species 변수 값별로 petal_width 상위 25% 값을 구한 후, 가장 큰 값과 가장 작은 값 사이의 차이를 계산하시오.

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/yemoonsaBigdata/datasets/Part2/iris.csv')
df.head()
```

```python
a = df.groupby(by='species')['petal_width'].quantile(0.75)
print(a)
```

```python
answer = a.max() - a.min()
print(answer)
```

#문제3
* iris 데이터 세트를 sepal_width 변수 값으로 내림차순한 후 위에서부터 100개 행을 학습 데이터로 분리하고, 학습 데이터에서 sepal_width의 표준편차 값을 구한 후 50을 곱한 값을 계산하시오(최종 계산 값의 소수점 이하는 버린다)

```python
df = pd.read_csv('/content/yemoonsaBigdata/datasets/Part2/iris.csv')

df1 = df.sort_values(by='sepal_width', ascending=False).reset_index(drop=True)

df_new = df1.loc[:99]

df_new.info()
```

```python
answer = np.trunc(df_new['sepal_width'].std()*50)
print(answer)
```

#문제4
* Iris 데이터 세트에서 species 변수 값별로 petal_length 상위 10개 행(큰 값)의 평균값을 구한 후, 평균값의 합을 계산하시오(최종 계산 값의 소수점 이하는 반올림한다.)

```python
df = pd.read_csv('/content/yemoonsaBigdata/datasets/Part2/iris.csv')

df = df.sort_values(by=['species','petal_length'], ascending=[True, False])

# 상위 10개 행 추출
df_new = df.groupby(by='species').head(10)
df_new
```

```python
df_new.groupby(by='species')['petal_length'].mean()
```

```python
answer = np.round(df_new.groupby(by='species')['petal_length'].mean().sum())

print(answer)
```

#문제5
* Iris 데이터 세트에서 sepal_length 변수가 이상치를 가지는 데이터 행 수를 계산하시오(단, 이상치 범위는 (평균+2*표준편차) 이상이거나 (평균-2*표준편차) 이하이다.)

```python
df = pd.read_csv('/content/yemoonsaBigdata/datasets/Part2/iris.csv')

cond1 = df['sepal_length'] >= df['sepal_length'].mean() + 2*df['sepal_length'].std()
cond2 = df['sepal_length'] <= df['sepal_length'].mean() - 2*df['sepal_length'].std()
```

```python
answer = len(df[cond1 | cond2])

print(answer)
```
