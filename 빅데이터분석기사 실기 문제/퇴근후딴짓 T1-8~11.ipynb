#T1-8 누적합 그리고 보간(결측치 처리)
* 주어진 데이터 셋에서 'f2' 컬럼이 1인 조건에 해당하는 데이터의 'f1'컬럼 누적합을 계산한다. 이때 발생하는 누적합 결측치는 바로 뒤의 값을 채우고, 누적합의 평균값을 출력한다. (단, 결측치 바로 뒤의 값이 없으면 다음에 나오는 값을 채워넣는다)
* 데이터셋 : basic1.csv


```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/basic1.csv')
```

```python
 # 'f2' 컬럼이 1인 조건에 해당하는 데이터의 'f1'컬럼 누적합을 계산
df[df['f2']==1]['f1'].cumsum()
```

```python
df_new = df[df['f2']==1]['f1'].cumsum()

df_new
```

```python
# 누적합 결측치는 바로 뒤의 값을 채움 -> method = 'bfill'
# 이전 값으로 채우려면 -> method = 'pad'
df_new = df_new.fillna(method='bfill')

df_new
```

```python
answer = df_new.mean()
print(answer)
```

#T1-10 수치형 변수 변환하기
* 주어진 데이터에서 20세 이상인 데이터를 추출하고 'f1'컬럼을 결측치를 최빈값으로 채운 후, f1 컬럼의 여-존슨과 박스콕스 변환 값을 구하고, 두 값의 차이를 절대값으로 구한다음 모두 더해 소수점 둘째 자리까지 출력(반올림)하시오
* 데이터셋 : basic1.csv

```python
df = pd.read_csv('/content/basic1.csv')
```

```python
df.head()
```

```python
# 20세 이상 데이터 추출
df_new = df[df['age'] >= 20]
df_new
```

```python
# f1컬럼 최빈값
a = df_new['f1'].mode()[0]

# 'f1'컬럼을 결측치를 최빈값으로 채우기
df_new = pd.DataFrame(df_new['f1'].fillna(a))

df_new
```

```python
# f1 컬럼의 여-존슨과 박스콕스 변환 값을 구하기
# sklearn.preprocessing 의 PowerTransformer 활용

from sklearn.preprocessing import power_transform

# yeo-johnson 변환값
df_new['y'] = power_transform(df_new[['f1']], standardize=False)

df_new['y']
```

```python
# box-cox 변환값
df_new['b'] = power_transform(df_new[['f1']], method='box-cox', standardize=False)

df_new['b']
```

```python
answer = round(sum(abs(df_new['y'] - df_new['b'])),2)

print(answer)
```

#T1-11 min-max 기준 상하위 5%값
* 주어진 데이터에서 'f5'컬럼을 min-max 스케일 변환한 후, 상위 5%와 하위 5% 값의 합을 구하시오
* 데이터셋 : basic1.csv

```python
df = pd.read_csv('/content/basic1.csv')
df.info()
```

```python
df['f5']
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df['f5'] = scaler.fit_transform(df[['f5']])
df['f5']
```

```python
high = df['f5'].quantile(0.95)

low = df['f5'].quantile(0.05)

answer = high + low

print(answer)
```
