#문제1
* Wine 데이터 세트에서 total_phenols 결측치를 제거하고 alcalinity_of_ash 변수의 결측치는 중앙값으로 대체한 후, alcalinity_of_ash 평균값을 계산하시오(최종 계산 값의 소수점 이하는 버린다.)

```python
import pandas as pd
import numpy as np

df = pd.read_csv('/content/yemoonsaBigdata/datasets/Part2/wine.csv')
df.head()
```

```python
df_drop = df[df['total_phenols'].notnull()]

df_drop.head()
```

```python
df_drop['alcalinity_of_ash'].fillna(df['alcalinity_of_ash'].median())
```

```python
answer = np.trunc(df['alcalinity_of_ash'].mean())

print(answer)
```

#문제2
* Wine 데이터 세트에서 alcohol 값이 alcohol 평균 값 이상이고, color_intensity 값이 color_intensity 평균 값 이상인 데이터만 남겼을 때, 가장 많은 class 값을 구하시오.

```python
df = pd.read_csv('/content/yemoonsaBigdata/datasets/Part2/wine.csv')

cond1 = df['alcohol'] >= df['alcohol'].mean()
cond2 = df['color_intensity'] >= df['color_intensity'].mean()

df_new = df[cond1 & cond2]

df_new.info()
```

```python
df_new['class'].unique()
```

```python
ans_pd = df_new['class'].value_counts()

print(ans_pd)
```

```python
type(ans_pd)
```

```python
answer = ans_pd.index[0]

print(answer)
```

#문제3
* Wine 데이터 세트에서 class 1에 해당하는 데이터의 ash 평균값과 Wine 데이터 세트의 모든 결측치를 제거한 후의 ash 값의 평균값의 차이의 절댓값을 계산하시오(최종 계산 값의 소수점 아래 4번째 자리에서 반올림한다.)

```python
df = pd.read_csv('/content/yemoonsaBigdata/datasets/Part2/wine.csv')

# class 1에 해당하는 데이터의 ash 평균값
df_1 = df[df['class']==1]
a = df_1['ash'].mean()
print(a)
```

```python
# Wine 데이터 세트의 모든 결측치를 제거한 후의 ash 값의 평균값

df_2 = df[df.notnull()]

b = df_2['ash'].mean()
print(b)
```

```python
answer = round(abs(a - b), 3)
print(answer)
```

#문제4
* Wine 데이터 세트에서 alcohol 변수의 상위에서 20번째 값(alcohol을 오름차순 정렬했을 때 20번째 위치한 값)으로 상위 20개의 값을 변환한 후, proanthocyanins가 2 이하인 데이터를 추출하여 alcohol의 평균값을 계산하시오(최종 계산 값의 소수점 아래 3번째 자리에서 반올림한다.)

```python
df = pd.read_csv('/content/yemoonsaBigdata/datasets/Part2/wine.csv')
df.info()
```

```python
df_new = df.sort_values(by='alcohol', ascending=False).reset_index(drop=True)
df_new.head()
```

```python
# alcohol 변수의 상위에서 20번째 값(alcohol을 오름차순 정렬했을 때 20번째 위치한 값)으로 상위 20개의 값을 변환
df_new.loc[:19,'alcohol'] = df_new.loc[19,'alcohol']
```

```python
answer = round(df_new[df_new['proanthocyanins'] <= 2]['alcohol'].mean(), 2)

print(answer)
```

#문제5
* Wine 데이터 세트에서 class 1에 해당하는 데이터의 proline 변수의 제3사분위수와 제1사분위수 차를 구하고, class 2에 해당하는 데이터의 proline 변수의 제3사분위수와 제1사분위수 차를 구하여 두 값의 차이의 절댓값을 계산하시오.

```python
df = pd.read_csv('/content/yemoonsaBigdata/datasets/Part2/wine.csv')

# Wine 데이터 세트에서 class 1에 해당하는 데이터의 proline 변수의 제3사분위수와 제1사분위수 차

a = df[df['class']==1]['proline'].quantile(0.25) - df[df['class']==1]['proline'].quantile(0.75)
```

```python
# class 2에 해당하는 데이터의 proline 변수의 제3사분위수와 제1사분위수 차

b = df[df['class']==2]['proline'].quantile(0.25) - df[df['class']==2]['proline'].quantile(0.75)
```

```python
answer = abs(a-b)
print(answer)
```
