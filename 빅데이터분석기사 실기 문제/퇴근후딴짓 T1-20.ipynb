* 고객과 잘 맞는 타입 추천 :)
* basic1 데이터와 basic3 데이터를 'f4'값을 기준으로 병합하고,
병합한 데이터에서 r2결측치를 제거한다음, 앞에서 부터 20개 데이터를 선택하고 'f2'컬럼 합을 구하시오
* basic1.csv: 고객 데이터
* basic3.csv: 잘 어울리는 관계 데이터 (추천1:r1, 추천2:r2)


```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('/content/basic1.csv')
df2 = pd.read_csv('/content/basic3.csv')

df1.head()
```

```python
df2.head()
```

```python
#df1과 df2를 'f4'컬럼 기준으로 병합

df = pd.merge(left=df1, right=df2, how='left', on='f4')

df.head()
```

```python
df.info()
```

```python
#r2 결측치 제거

df = df.dropna(subset='r2')
df.info()
```

```python
df = df.head(20)
answer = df['f2'].sum()

print(answer)
```
