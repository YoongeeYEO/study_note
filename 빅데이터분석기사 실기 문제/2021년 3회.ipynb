# 작업형 1유형 - 02
 * [연도별 나라별 유병률 데이터] 데이터 중 2000년도에 전체 국가 유병률의 평균보다 큰 나라 수를 구하라.

```python
data3 = pd.read_csv('/content/yemoonsaBigdata/datasets/Part3/302_worlddata.csv')

data3
```

```python
data3_new = data3[data3['year']==2000]

data3_new
```

```python
data3_new = data3_new.drop(['year'], axis=1)
```

```python
data3_new = data3_new.T

data3_new
```

```python
baseline = data3_new[1].mean()

print(baseline)
```

```python
answer = len(data3_new[data3_new[1]>baseline])

print(answer)
```
