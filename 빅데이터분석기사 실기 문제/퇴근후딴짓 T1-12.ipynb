#T1-12. 상위 10개, 하위 10개 차이 Expected Questions
* 문제 : 주어진 데이터에서 상위 10개 국가의 접종률 평균과 하위 10개 국가의 접종률 평균을 구하고, 그 차이를 구해보세요

  (단, 100%가 넘는 접종률 제거, 소수 첫째자리까지 출력)

  데이터셋 : ../input/covid-vaccination-vs-death/covid-vaccination-vs-death_ratio.csv

```python
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/빅분기 실기 연습문제/covid-vaccination-vs-death_ratio.csv')

data.head()
```

```python
df_new = data.groupby('country').max()
df_new = df_new.sort_values(by='ratio', ascending=False)

df_new.head()
```

```python
df_new = df_new[df_new['ratio']<=100]

df_new.head()
```

```python
top10 = df_new['ratio'].head(10).mean()
tail10 = df_new['ratio'].tail(10).mean()


import numpy as np
answer = np.round(top10 - tail10,1)
print(answer)
```
