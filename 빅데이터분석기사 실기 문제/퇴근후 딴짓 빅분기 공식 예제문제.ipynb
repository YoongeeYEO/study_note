#문제1
* 자동차 데이터 셋에서 qsec 컬럼을 Min-Max Scale로 변환 후 0.5보다 큰 값을 가지는 레코드(row) 수를 구하시오

```python
import pandas as pd

df = pd.read_csv('/content/mtcars.csv')
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import minmax_scale

df['qsec'] = minmax_scale(df['qsec'])
```

```python
df['qsec']
```

```python
answer = len(df[df['qsec'] > 0.5 ])

print(answer)
```

* MinMaxScaler 활용

```python
df = pd.read_csv('/content/mtcars.csv')

df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

df['qsec'] = MinMaxScaler(df['qsec'])

df['qsec']
```

```python
answer = len(df[df['qsec'] > 0.5 ])

print(answer)
```

* MinMaxScaler() 에서는 minmax_scale()과 다르게 에러 발생하는데 이유는?
