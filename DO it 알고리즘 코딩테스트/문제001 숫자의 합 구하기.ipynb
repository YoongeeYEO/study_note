# 문제 001 : 숫자의 합 구하기


- N개의 숫자가 공백 없이 써 있다. 이 숫자를 모두 합해 출력하는 프로그램을 작성하시오.

- 입력 : 1번째 줄에 숫자의 갯수(1<=N<=100), 2번째 줄에 숫자 N개가 공백 없이 주어진다.

- 출력 : 입력으로 주어진 숫자 N개 합을 출력한다.

```python
n = int(input())

numbers = list(input())

sum = 0

for i in numbers:
    sum = sum + int(i)
print('입력된 숫자의 총합은',sum,'입니다.')
```

숫자의 갯수만큼 입력받은 값을 리스트형태로 저장하면 입력한 숫자의 각 값이 순서대로 리스트에 저장된다.

리스트 내 각 값을 Int 형태로 변환하고 결과값에 더하여 누적한 sum 값을 출력한다.
