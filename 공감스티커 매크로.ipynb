* pyautogui 다운로드 후 import 하고 사용중인 모니터 해상도 & 현재 마우스 커서 위치를 출력할 수 있다.

```python
import pyautogui

# 현재 사용하는 모니터의 해상도 출력
print(pyautogui.size())

# 현재 마우스 커서의 위치 출력
print(pyautogui.position())
```

```python
#맥북 화면에서 카카오톡 앱 1번 클릭
time.sleep(1)
pyautogui.click(x=873, y=941, clicks=1, button='left')

#카카오톡 프로필 1번 클릭
time.sleep(1)
pyautogui.click(x=181, y=241, clicks=1, button='left')

#공감스티커 위치로 가서 왼쪽클릭 1000번 수행
time.sleep(1)
pyautogui.click(x=462, y=345, clicks=1000, interval=0.2, button='left')
```

* 재미로 한거라, 절대위치를 기준으로 마우스 커서를 옮겨서 왼쪽클릭 하는것만 간단하게 해보았는데 pyautogui 라이브러리를 사용하면 마우스 뿐 아니라 숫자/영어문장 입력하기, 차례대로 키입력, 조합키 입력하기 (ex : shift + 4, ctrl + a), 한글 문자열 입력하기 (복사 붙여넣기 이용), 화면에 이미지 존재하는지 확인 (locateOnScreen) 등등 다양한 gui 작업들을 자동화시킬 수 있다.


* Window, macOS, Linux등의 대중적 운영체제에 모두 적용가능함.
